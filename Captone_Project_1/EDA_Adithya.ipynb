Playstore data

The variables Reviews, Size, Installs and Price are of object data type. We need to convert it into float or integer data type to work with it.

```python
# Converting 'Reviews' variable from the type 'object' to 'float'
reviews = [i for i in playstore_data['Reviews']]

def into_float1(reviews_list):
    """
    converting 'Reviews' variable 
    from the type 'object' to 'float'
    """
    cleaned_review = []
    for review in reviews_list:
        if 'M' in review:
            review = review.replace('M', '')
            review = float(review) * 1000000  # 1M = 1,000,000
        cleaned_review.append(review)
    return cleaned_review

playstore_data['Reviews'] = into_float1(reviews)
playstore_data['Reviews'] = playstore_data['Reviews'].astype(float)
```

```python
# We can observe that size variable has one value mentioned 1000+, which is impractical. So initially we will remove that row and then we will convert it's data type into float.
index = playstore_data[playstore_data['Size'] == '1000+'].index
playstore_data.drop(index = index , inplace = True,axis = 0)

size = [i for i in playstore_data['Size']]
def into_float2(size_list):
  """ Converting datatype of size variable into 
  float """
  cleaned_size = []
  for i in size_list:
    if 'M' in i:
      i = i.replace('M', '')
      i = float(i)
    elif 'k' in i:
      i = i.replace('k', '')
      i = float(i)
      i = i/1024   # As 1mb = 1024kb
    else:
      i = float(0)
    cleaned_size.append(i)
  return cleaned_size

playstore_data['Size'] = into_float2(size)
playstore_data['Size'] = playstore_data['Size'].astype(float)
```

```python
# Observe that the variable installs is of object datatype. We will convert it into float datatype.
index = playstore_data[playstore_data['Installs'] == 'Free'].index
playstore_data.drop(index = index , inplace = True,axis = 0)
installs = [i for i in playstore_data['Installs']]

def into_float3(install_list):
  """ Converting datatype of installs
  into float """
  cleaned_installs = []
  for i in install_list:
    if ',' in i:
      i = i.replace(',', '')
    if "+" in i:
      i = i.replace('+', '')
    i = float(i)
    cleaned_installs.append(i)
  return cleaned_installs

playstore_data['Installs'] = into_float3(installs)
playstore_data['Installs'] = playstore_data['Installs'].astype(float)
```

```python
# Converting datatype of Price into float.
price = [i for i in playstore_data['Price']]

def into_float4(price_list):
  """ Converting datatype of
  Price into float. """
  cleaned_price = []
  for i in price_list:
    if '$' in i:
      i  = i.replace("$", '')
      i = float(i)
    cleaned_price.append(i)
  return cleaned_price

playstore_data['Price'] = into_float4(price)
playstore_data['Price'] = playstore_data['Price'].astype(float)
```

```python
# 8.Let us group the applications based on Size of application.
def size_apps(data):
  '''
  This function group the applications into
  certain intervals based on the size of 
  the applications.
  '''
  try:
    if data < 1:
      return 'Below 1'
    elif data >= 1 and data <10:
      return '1-10'
    elif data >= 10 and data <20:
      return '10-20'
    elif data >= 20 and data <30:
      return '20-30'
    elif data >= 30 and data <40:
      return '30-40'
    elif data >= 40 and data <50:
      return '40-50'
    elif data >= 50 and data <60:
      return '50-60'
    elif data >= 60 and data <70:
      return '60-70'
    elif data >= 70 and data <80:
      return '70-80'
    elif data >= 80 and data <90:
      return '80-90'
    else:
      return '90 and above'
  except:
    return data
```

```python
playstore_data['size_group']=playstore_data['Size'].apply(lambda x : size_apps(x))
playstore_data.head()
```
